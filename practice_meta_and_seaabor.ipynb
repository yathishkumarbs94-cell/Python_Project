# 4. Handling Outliers

```python
# Q1. Define thresholds using IQR method - ratings_count
q1 = data['ratings_count'].quantile(0.25)
print(q1)

q3 = data['ratings_count'].quantile(0.75)
print(q3)

iqr = q3 -q1
print(iqr)

upper_threshold = q3 + (1.5 * iqr)
lower_threshold = q1 - (1.5 * iqr)
print(upper_threshold,'&' ,lower_threshold)
```

```python
# Q2. Flag outliers for further review
# Create a new column is_outlier_rating for outliers in ratings_count.
# New columns should have True if value > upper threshold or value < lower threshold

condition=(data['ratings_count']>upper_threshold) | (data['ratings_count'] < lower_threshold)
data['ratings_count_outlier_flag'] = condition
```

```python
data['ratings_count_outlier_flag_2'] = data['ratings_count'].apply(lambda x: True if (x> upper_threshold) or (x < lower_threshold ) else False )
```

```python
#data.head(100)

# Outlier Treatment
# Dropping the outlier --> Outliers are wrong data
# Create a flag column based on IQR threshold for further analysis --> True or False | further analysis we can over this data (Majority in the case of analytics)
# Capping / Flooring the outlier --> data > upper threshold --> upper_threshold , data < lower threshold --> lower_threshold (Data Cleaning --> ML Models)
```

# String Cleaning

```python
# Q1. Strip whitespace from title and authors.
# Method 1
data['title']=data['title'].str.strip()
data['authors'] = data['authors'].str.strip()
# Method 2
data['title'] = data['title'].apply(lambda x : x.strip())
data['authors'] = data['authors'].apply(lambda x : x.strip())
```

```python
# Q2. Convert authors to title case.
# 'I am Leaning Data science'.title()
data['authors'] = data['authors'].apply(lambda x : x.title())
```

```python
data.head()
```

```python
# Q3. Remove special characters from title.
def remove_spec(text):
  for i in['$','#',':','*','@','^']:
    return text

  data['title']=data['title'].apply(remove_spec)
```

```python
data['title'].head(5)
```

```python
i = 'a1'
# Scalable method --> regex
i.isalnum()

def remove_special_characters(text):
  new_text = ''
  for i in text:
    if i.isalnum() or i in [' ', ':']:
      new_text+= i
  return new_text

text = 'Tyrannosaurus Wrecks (Stanley #1)'
remove_special_characters(text)
```
